```python
def print_n(s="Hello", n = 2):
    if n <= 0:
        return
    print(s)
    print_n(s, n-1)

#print_n("Hello", 2)

def do_n(func, n):
  if n <= 0:
    return
  else:
    func()
    do_n(func, n-1)
    
do_n(print_n, 7)
```

```python
def get_day_of_the_week(day=0):
  if day == 0:
    return 'Monday'
  elif day == 1:
    return "Tuesday"
  elif day == 2:
    return 'Wednesday'
  elif day == 3:
    return 'Thursday'
  elif day == 4:
    return 'Friday'
  elif day == 5:
    return 'Saturday'
  elif day == 6:
    return 'Sunday'
  else:
    return "BAD INPUT"
```

```python
def factorial_loop(n):
  product = 1
  while n >= 1:
    product *= n
    n -= 1
  return product
```

```python
factorial_loop(5)
```

```python
def factorial_recursive(n):
  if n == 1:
    return 1
  else:
    return n * factorial_recursive(n-1)
```

```python
factorial_recursive(5)
```

```python
def bad_factorial(n):
  print(n)
  return n * bad_factorial(n-1)

  if n == 1:
    return 1
```

```python
def fibonacci(n):
  print('Computing fibonacci(n) for n =', n)
  if (n >= 0) and (n < 2):
    return n
  return fibonacci(n - 1) + fibonacci(n - 2)
```

```python
def fibonacci_memo(n, memo={0:0, 1:1, 2:1}):
  print('Computing fibonacci_memo(n) for n =', n)
  if n in memo:
    return memo[n]
  #if (n >= 0) and (n < 2):
  #  memo[n] = n
  #  return memo[n]
  memo[n] = fibonacci_memo(n - 1, memo) + fibonacci_memo(n - 2, memo)
  return memo[n]
```
