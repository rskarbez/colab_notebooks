```python
australian_lgas = gpd.read_file('LGA_2022_AUST_GDA2020.shp')
```

```python
australian_lgas.plot()
```

```python
aus_axes = gplt.polyplot(australian_lgas, projection=gplt.crs.Mercator(), figsize=(25, 15))

australian_rail.plot(ax=aus_axes)
```

```python
base = australian_lgas.plot(color='white', edgecolor='black')

australian_rail.plot(ax=base, color='red');
```
