```python
input = [['Harry', 37.21], ['Berry', 37.21], ['Tina', 37.2], ['Akriti', 41], ['Harsh', 39]]
```

```python
# function to enable sorting by second index : test scores     
def takeSecond(elem):
  return elem[1] 
```

```python
def ruperts_fcn(input):
  class_scores = []
  for name, score in input:
    #name = input()
    #score = float(input())
    #adding student name and test score to nested lists
    class_scores.append([name, score])
        
  class_scores.sort(key=takeSecond) #sorting list by test scores low to high

  lowest = class_scores[0][1] # lowest score value 

  #setting second_lowest to 100 until find value above 'lowest'
  second_lowest = 100
  second_low_students = []

  for s in class_scores:
    if s[1] > second_lowest:
      break
    # catching all student names whose score is immediately above 'lowest' score
    if s[1] > lowest: 
      # setting second_lowest to this value so don't getting any names with score higher than this
      second_lowest = s[1] 
      second_low_students.append(s[0])
        
  second_low_students.sort()
  for s in second_low_students:
    print(s)
```

```python
%%timeit
ruperts_fcn(input)
```

```python
def editorial_fcn(input):
  score_list = {}
  for name, score in input:
    if score in score_list:
        score_list[score].append(name)
    else:
        score_list[score] = [name]
  new_list = []
  for i in score_list:
    new_list.append([i, score_list[i]])
  new_list.sort()
  result = new_list[1][1]
  result.sort()
  print (*result, sep = "\n")
```

```python
%%timeit
editorial_fcn(input)
```
