```python
def fibonacci(n):
  if n == 0:
    #print('Computing fibonacci(0)...')
    return 0
  elif n == 1:
    #print('Computing fibonacci(1)...')
    return 1
  else:
    #print('Computing fibonacci({})...'.format(n))
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
import time
help (time.perf_counter)
```

```python
%%timeit
#tick = time.perf_counter()
fibonacci(30)
#tock = time.perf_counter()
#print(tock - tick)
```

```python
def fibonacci_memo(n, memo={0:0, 1:1}):
  if n in memo:
    return memo[n]
  else:
    memo[n] = fibonacci_memo(n-1, memo) + fibonacci_memo(n-2, memo)
    return memo[n]
```

```python
%%timeit
fibonacci_memo(300)
```
