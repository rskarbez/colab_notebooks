```python
import pandas
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
tips
```

```python
tips_sun = tips[tips['day'] == 'Sun']
print(tips_sun)
```

```python
tips_thur = tips[tips['day'] == 'Thur']
tips_thur
```

```python
plt.scatter(tips_sun['total_bill'], tips_sun['tip'] / tips_sun['total_bill'] * 100, c='orange')
plt.scatter(tips_thur['total_bill'], tips_thur['tip'] / tips_thur['total_bill'] * 100, c='blue')
plt.show()
```
